```python
# Assignment no 16

import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

# Recall actual file paths
file_path1 = r"C:\Users\munir\Downloads\Muni_energy-check-argon-nve-rlist09.dat"
file_path2 = r"C:\Users\munir\Downloads\Muni_energy-check-argon-nve-rlist10.dat"

# Read data from the CSV files into DataFrames
data1 = pd.read_csv(file_path1, delimiter='\s+', header=None, names=['Time', 'Energy'])
data2 = pd.read_csv(file_path2, delimiter='\s+', header=None, names=['Time', 'Energy'])

# Plotting
plt.plot(data1['Time'], data1['Energy'], label='measurement 1', color='blue', marker='o')
plt.plot(data2['Time'], data2['Energy'], label='measurement 2', color='red', marker='s')

# Linear Regression
model = LinearRegression()

# Fit the model for File 1
model.fit(data1[['Time']], data1['Energy'])
plt.plot(data1['Time'], model.predict(data1[['Time']]), linestyle='--', color='blue')

# Fit the model for File 2
model.fit(data2[['Time']], data2['Energy'])
plt.plot(data2['Time'], model.predict(data2[['Time']]), linestyle='--', color='red')

# Customize the plot
plt.title('GROMACS Energies and Linear Regression Fits')
plt.xlabel('Time (ps)')
plt.ylabel('Total Energy (kJ/mol)')
plt.legend()
plt.grid(True)

# Save the plot as an image (optional)
plt.savefig('comparison_plot_with_regression.png')

# Display the plot
plt.show()
```

```python
# Assignment no 18
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np

# Recall actual file paths
file_path1 = r"C:\Users\munir\Downloads\Muni_energy-check-argon-nve-rlist09.dat"
file_path2 = r"C:\Users\munir\Downloads\Muni_energy-check-argon-nve-rlist10.dat"

# Read data from the CSV files into DataFrames
data1 = pd.read_csv(file_path1, delimiter='\s+', header=None, names=['Time', 'Energy'])
data2 = pd.read_csv(file_path2, delimiter='\s+', header=None, names=['Time', 'Energy'])


# Assuming data2 is your DataFrame, and 'Column1' is the column of interest
fluctuations_data2 = data2['Energy']

# Plot histogram
plt.hist(fluctuations_data2, bins=30, density=True, alpha=0.6, color='g')

# Fit a normal distribution to the data
mu, std = norm.fit(fluctuations_data2)

# Plot the fitted Gaussian
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)

# Add labels and legend
plt.title('Histogram of Fluctuations for Data2')
plt.xlabel('Fluctuations')
plt.ylabel('Frequency')
plt.legend(['Fit ($\mu={:.2f}$, $\sigma={:.2f}$)'.format(mu, std), 'Data'])

# Show the plot
plt.show()
```
